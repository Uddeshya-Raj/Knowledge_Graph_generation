```python
import json
```

```python
BOOK_NAME = "Bharatiya Vratotsav"
N = 10 # no. of chapters to process
```

```python

chunks = {}
counter = 1


for i in range(N):
    with open(f"../entity_output_files_redone/chapter_{i}_output.json", "r", encoding="utf-8") as f:
        file = json.load(f)
        for chunk in file:
            entities_in_chunk = set()
            sentences = []
            for obj in chunk['extracted_entities']:
                entities_in_chunk.update(obj['entities'])
                sentences.append(obj['sentence'])
            chunks[f"chunk_{counter}"] = {
                "node_id": f"chunk_{counter}",
                "book": BOOK_NAME,
                "chapter": i+1,
                "content": '\n'.join(sentences),
                "entities_present": list(entities_in_chunk),
                "prev_chunk": f"chunk_{counter-1}" if counter-1 else None,
                "next_chunk": f"chunk_{counter+1}"
            }
            counter += 1

    chunks[f"chunk_{counter-1}"]["next_chunk"] = None
    
        
```

```python
len(chunks)
```

```python
with open("../GraphRAG_structure_setup/chunks.json", "w", encoding="utf-8") as f:
    json.dump(chunks, f, indent=2, ensure_ascii=False)
```
